```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
ipl_df = pd.read_csv('ipl.csv')
```

```python
ipl_df.head(3)
```

```python
tot_win_by_runs = ipl_df[['winner','win_by_runs']]
tot_win_by_runs = tot_win_by_runs.groupby('winner').sum()
```

```python
sns.barplot(x='winner',y='win_by_runs',data = tot_win_by_runs);
```

```python
std_season = ipl_df.groupby(['season','city'])[['id']].count()
std_season
plt.figure(figsize=(12, 20))
sns.barplot(x='season', y='id', hue='city', data=std_season)

plt.xlabel('Season')
plt.ylabel('Match Count')
plt.title('Number of Matches by City and Season')
```

```python
data = {
    'season': [2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022],
    'city': ['CityA', 'CityB', 'CityC', 'CityA', 'CityB', 'CityC', 'CityA', 'CityB', 'CityC'],
    'id': [10, 15, 12, 8, 12, 14, 9, 11, 13]
}
df = pd.DataFrame(data)

plt.figure(figsize=(4, 3))
sns.barplot(x='season', y='id', hue='city', data=df)
plt.xlabel('season')
plt.ylabel('count')
```

```python
tips = sns.load_dataset("tips")
# Create a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips)

# Add title and labels
plt.title('Boxplot of Total Bill by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.grid(True)

# Show the plot
plt.show()
```
