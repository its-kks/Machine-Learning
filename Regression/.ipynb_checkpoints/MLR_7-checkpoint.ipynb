```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm 
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
"""
    REMEMBER
"""
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(3)
```

```python
df = pd.get_dummies(df,drop_first=True)
X = df.drop('tip',axis=1)
Y = df['tip']
```

```python
# Removing Multicollinearity
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
# Replacing True and False
for col in X.columns:
    X[col] = X.apply(lambda rec : rec[col] if rec[col] != True else 1, axis=1)
    X[col] = X.apply(lambda rec : rec[col] if rec[col] != False else 0, axis=1)

"""
    REMEMBER 
    1. A VIF of 1 indicates no multicollinearity,
    15 suggests moderate but acceptable multicollinearity, 
    >5 signals potential issues, and 
    >10 indicates severe multicollinearity requiring corrective action.
"""
# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(X.shape[1])]

print(vif_data.sort_values('VIF',ascending=False))

mask = vif_data['VIF'] >= 10
col_to_remove = vif_data[mask]['feature']
X = X.drop(col_to_remove,axis=1)
X = sm.add_constant(X)
print(X.head())
```

```python
train_X, test_X, train_y,test_y = train_test_split(X,Y,random_state=42,train_size=0.8)

OLS_model = sm.OLS(train_y,train_X).fit()

print(f"Coefficients:\n{OLS_model.params}")
print(f"Rsqured:{OLS_model.rsquared}")
print(f"P_Values:\n{OLS_model.pvalues}")
```

```python
pred_df = pd.DataFrame()
pred_df['pred_tip'] = OLS_model.predict(test_X)
pred_df['actual_tip'] = test_y
print(pred_df.head(3))
```

```python
print(f"Rsquare: {r2_score(test_y, pred_df['pred_tip'])}")
print(f"MSE: {mean_squared_error(test_y,pred_df['pred_tip'])}")
```

```python
"""
    REMEMBER 
    1. Residulas = actual - predicted
    2. Scatter plot x : predicted y : residual
    3. A funnel shape in a residual plot indicates heteroscedasticity ( variance of error is not constant),
    which is the opposite of homoscedasticity
"""

pred_df['residuals'] = pred_df['actual_tip'] - pred_df['pred_tip']

mean_res = pred_df['residuals'].mean()
std_res = pred_df['residuals'].std()
pred_df['residuals_stand'] = pred_df.apply(lambda rec: (rec['residuals'] - mean_res) / std_res, axis=1)

mean_actual = pred_df['pred_tip'].mean()
std_actual = pred_df['pred_tip'].std()

pred_df['pred_tip_stand'] = pred_df.apply(lambda rec: (rec['pred_tip'] - mean_actual) / std_actual, axis=1)

plt.scatter(x=pred_df['pred_tip_stand'], y=pred_df['residuals_stand'])
```
