```python
kks_df = { 'name' : ['KKS','FKS','JKS'] , 'cost' : [50,60,70]}
kks_df = pd.DataFrame(kks_df)
```

```python
rand_values = np.array([ rand.randint(10,30) for _ in range(kks_df.shape[0]) ])
print(rand_values)
```

```python
kks_df['rand'] = rand_values
kks_df
```

```python
rand_values.reshape(3,1)
kks_df['rand2'] = rand_values
kks_df
```

```python
kks_df
```

```python
kks_df['sum_rand'] = kks_df.apply(lambda rec : rec['rand'] + rec['rand2'] , axis=1)
kks_df
```
